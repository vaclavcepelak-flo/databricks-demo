```python
from ydata_profiling import ProfileReport
```

```python
df = spark.table("sandbox_raw.bikesharing.day")
```

```python
df.count()
```

```python
display(df)
```

```python
# ProfileReport may not work on serverless compute due to RDD limitations.
# Consider using mapInPandas or mapInArrow for custom transformations.
report = ProfileReport(df.toPandas(),
                title='Bike Rental Daily Data',
                infer_dtypes=False,
                interactions=None,
                missing_diagrams=None,
                correlations={"auto": {"calculate": False},
                              "pearson": {"calculate": True},
                              "spearman": {"calculate": True}})
```

```python
report_html = report.to_html()
displayHTML(report_html)
```
